```python
wal_laptop_data = {}
for index, row in wal_laptop_df.iterrows():
   wal_laptop_data[row["upc"]]=row

compare_data = []
for index, row in types_df.iterrows():
    query = row["upc"]
    if query in wal_laptop_data:
        #if row["customerReviewCount"]=0
        new_data = {
           "upc":query,
           "BestBuy_SalePrice":row["salePrice"],
           "BestBuy_ReviewCount":row["customerReviewCount"],
           "BestBuy_ReviewAverage":row["customerReviewAverage"],
           "Walmart_SalePrice":wal_laptop_data[query]["salePrice"],
           "Walmart_ReviewCount":wal_laptop_data[query]["numReviews"],
           "Walmart_ReviewAverage":wal_laptop_data[query]["customerRating"]
        }
        compare_data.append(new_data)
        #print(wal_laptop_data[query])
       
compare_pd = pd.DataFrame(compare_data)
compare_pd=compare_pd.fillna(0)
compare_pd.to_csv("combined_datasets/bestBuy_walmart_compare_data.csv", sep=',')
compare_pd
```
