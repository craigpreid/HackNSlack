```python
# Dependencies
import requests
import json
import csv
import pandas as pd
import time
```

```python
url = "https://api.bestbuy.com/v1/products((categoryPath.id=pcmcat209400050001))?apiKey=Jz4AJitfYbGIGGaOuTR09Lpl&sort=customerReviewAverage.asc&show=customerReviewAverage,name,customerReviewCount,inStoreAvailability,onlineAvailability,regularPrice,salePrice,upc,manufacturer&pageSize=100&page="
```

```python
phone_list = []


for page in range(1, 9):

    # create query with page number
    query_url = url + str(page) + "&format=json"

    products = requests.get(query_url).json()
    print(query_url)
    # Add a one second interval between queries to stay within API query limits
    time.sleep(1)
    
    # loop through the response and append each article to the list
    for product in products["products"]:
        phone_list.append(product)
```

```python
print(phone_list)

print(json.dumps(phone_list, indent=4, sort_keys=True))
```

```python
phonebestbuy_df = pd.DataFrame(phone_list)
phonebestbuy_df.head()
```

```python
phonebestbuy_df.to_csv("PhoneBestbuy.csv", index=False, header=True)
```
