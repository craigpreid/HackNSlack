```python
# Dependencies
import requests
import json
import csv
import pandas as pd
```

```python
#url to pull laptop data

url="https://api.bestbuy.com/v1/products((categoryPath.id=abcat0502000))?apiKey=Jz4AJitfYbGIGGaOuTR09Lpl&sort=manufacturer.asc&show=manufacturer,name,onlineAvailability,upc,regularPrice,salePrice,customerReviewAverage,customerReviewCount&pageSize=100&format=json"
```

```python
print(requests.get(url))
```

```python
# Pretty Print the output of the JSON
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
laptop_list=response['products']
```

```python
laptopbestbuy_df = pd.DataFrame(laptop_list)
laptopbestbuy_df.head()
```

```python
laptopbestbuy_df.to_csv("LaptopBestbuy.csv", index=False, header=True)
```
