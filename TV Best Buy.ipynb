```python
# Dependencies
import requests
import json
import csv
import pandas as pd
import time
import re
```

```python
url = "https://api.bestbuy.com/v1/products((categoryPath.id=abcat0101000))?apiKey=Jz4AJitfYbGIGGaOuTR09Lpl&sort=customerReviewAverage.asc&show=customerReviewAverage,name,customerReviewCount,inStoreAvailability,onlineAvailability,regularPrice,salePrice,upc,manufacturer&pageSize=100&page="
```

```python
tv_list = []


for page in range(1, 4):

    # create query with page number
    query_url = url + str(page) + "&format=json"

    products = requests.get(query_url).json()
    print(query_url)
    # Add a one second interval between queries to stay within API query limits
    time.sleep(1)
    
    # loop through the response and append each article to the list
    for product in products["products"]:
        product["screensize"] = int(re.findall(r'[0-9\.]+"', product["name"])[0].replace("\"",""))
        print(product)

        tv_list.append(product)
```

```python
print(tv_list)

print(json.dumps(tv_list, indent=4, sort_keys=True))
```

```python
tvbestbuy_df = pd.DataFrame(tv_list)
tvbestbuy_df.head()
```

```python
tvbestbuy_df.to_csv("TVBestbuy_craig.csv", index=False, header=True)
```

```python
bins = [0, 50, 60, 70, 75,105]
group_names = ["0 to 50 inches", "50 to 60 inches", "60 to 70 inches","70 to 75 inches","Greater than 75 inches"]
tvbestbuy_df["TV Size Category"] = pd.cut(tvbestbuy_df["screensize"], bins, labels=group_names)
tvbestbuy_df
```

```python
tvbestbuy_df

sizegroup = tvbestbuy_df["TV Size Category"].value_counts()
sizegroup_df = pd.DataFrame(sizegroup)
sizegroup_df.head()
```

```python
grouped_df = tvbestbuy_df.groupby(['TV Size Category'])
print(grouped_df)
grouped_df.head()
```

```python

averageprice = grouped_df["regularPrice"].mean()
```
